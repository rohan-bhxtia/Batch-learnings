 # important modes 

### r = reading [default]
### w = writing, truncate file first ( mtlb sb remove krke jo likhoge woh likhdega )
### x = creates new and open for writing ( nayi file bnane keliye)
### a = writing , appends at end 
### b = binary mode 
### t = text mode [default]
### + = oepn disk file for updater (r & w)



```python
f = open("sample.txt","r")
print(f.read())
f.close()
```

```python
f = open("sample.txt","w")
f.write("jai kara shera wali da")
f.close()
```

```python
f = open("sample.txt","a")
f.write("\nbole sache darbar di jaiiii")
f.close()
```

```python
f = open("sample2.txt","x")
f.write("jai shree krishna")
f.close()
```

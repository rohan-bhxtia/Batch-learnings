```python
salary = float(input("Enter your salary: "))

if salary < 30000:
    tax_rate = 5
elif salary <= 70000:
    tax_rate = 15
else:
    tax_rate = 25

tax_amount = salary * tax_rate / 100

print(f"Salary: {salary}")
print(f"Tax Rate: {tax_rate}%")
print(f"Tax Amount: {tax_amount}")
```

```python
## sum of digits
def numbers(a):
  sum = 0
  while a>0:
    last_digit = a%10
    sum += last_digit
    a = a//10
  print(sum)

a = int(input("enter a : "))

numbers(a)
```

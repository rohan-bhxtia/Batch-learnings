```python
# Q3. Create a program that:
# 1. Has a list of numbers: [5, 10, 15, 20, 25]
# 2.Uses a list comprehension to create a new list with only numbers greater than 15
# 3.Prints the new list

nums = [5,10,15,20,25]
new_lst = [i for i in (nums) if i>15] 
print(new_lst)
```

```python
# Create a Python dictionary of 3 cities and their populations. Save it to "cities.json".
# 1. Then load the JSON and print each city and its population.
# 2.Ask the user for a new city & its population - update this info in the json file.
import json
with open("cities.json","w") as f:
    data = {
            "delhi" : 100,
            "bathinda" : 50,
            "goniana" : 10
    }
    json.dump(data,f,indent = 3)    

        
```

```python
import json

with open("cities.json","r") as f :
    data = json.load(f)

for cities, population in data.items():
    print(cities,":",population)    
```

```python
import json 
with open ("cities.json","r") as f:
    data = json.load(f)
    new_city = str(input("enter city name:"))
    city_population = int(input("enter the city's popularion"))

    data[new_city] = city_population

with open("cities.json","w") as f:
    json.dump(data,f,indent = 3)


print("city added successfully ")
with open("cities.json","r") as f :
    print(f.read())
```

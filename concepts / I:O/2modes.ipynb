```python
# r = reading mode [ddefault]

# w = writing mode 

# x = creates new and opens for writing 

# a = writing , append at end 
# b = binary mode 

# t = text mode 
# +  = opens disk file for updater ( r & w) r+ , w+ ,a+ 

#we can also use multple modes at same time , like 
#rt(read text ) , rb (read binary ) , wt.  , wb 
```

```python
# read
f = open("sample.txt","r")
data = f.read()
print(data)
f.close()
```

```python
# write 
f = open("sample.txt","w")
f.write("im the best")
f.close()
```

```python
# new file 
f = open("sample.txt2","x")
f.write("this is appended file with mode X")
f.close()
```

```python
# append 
f = open("sample.txt2","a")
f.write("\nand im best too")
f.close()
```

```python
# r+ ( means we will read first and then write )
# print hoga pehle text , then add hojayega ( internally ) , added text print nhi hoga  

f = open("sample.txt","r+")
f.write("123")
print(f.read())
f.close()
```

```python
# w+ ( means pehle sb truncate(delete) hojayega then write hoga )

f = open("sample.txt","w+")
f.write("hehe 123")
print(f.read())
f.close()
```

```python
# a+ (pehle append hoga last mein )

f = open("sample.txt","a+")
f.write("appended text")
print(f.read())

f.close()
```

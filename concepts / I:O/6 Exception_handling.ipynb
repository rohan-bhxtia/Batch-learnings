## following are the 4 keywords to use :
### 1.try ( code line where the chances of error are)
### 2.except ( write the exception name u want to display)
### 3.else ( what do u want to print if excpetion wasnt there )
### 4.finally ()

```python
try :
    x = int(input("enter x: "))
    ans =10/x
    
except ZeroDivisionError:
    print("divisible by 0 is not allowed")        
else:
    print(ans)    
```

```python
try :
    x = int(input("enter x: "))
    ans =10/x
except ZeroDivisionError:
    print("divisible by 0 is not allowed")        
except ValueError:
    print("no characetrs allowed , only integers except 0")    
else:
    print(ans)    
```

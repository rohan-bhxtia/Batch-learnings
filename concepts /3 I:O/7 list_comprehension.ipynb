```python
# print a list including squares of numbers from 1 to 5
squares = []
for i in range(6):
    squares.append(i*i)
print(squares)
```

```python
sq = [i*i for i in range(6)]
print(sq)
```

### with conditions

```python
sq = [i*i for i in range(6) if i%2==1]
print(sq)
```

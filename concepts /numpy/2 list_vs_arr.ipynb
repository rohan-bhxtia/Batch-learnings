```python
import numpy as np 
import time 
```

```python
# we'll perform execution and see whos faster 

size = 100_000_000

py_list = list(range(size))
start = time.time()
square = [x*x for x in py_list ]
end = time.time()
print(f"python list time for execution = {end - start} seconds")

```

# VECTORIZATION 
## vectorization is the concept where operations on entire arrays without explicit loops , making code cleaner and faster 

```python
np_arr = np.array(py_list)
start = time.time()
sq_arr = np_arr ** 2 
end = time.time()
print(f"numpy array time = {end-start} seconds") 
```

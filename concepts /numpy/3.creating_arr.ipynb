# operations on arrays

## 1.reshaping 

```python
t_arr = np.array([[1,2,3],[4,5,6]])
print(t_arr,t_arr.shape)
reshape = t_arr.reshape((3,2))
print(reshape,reshape.shape)
```

```python
reshape2 = t_arr.flatten() # to convert to 1D array u need to flatten 
print(reshape2,reshape2.shape)
```

## indexing 

```python
arr = np.array([1,2,3,4,5])
print(arr[2])
```

```python
print(arr[arr>2]) ## boolean indexing 
```

```python
print(arr[arr%2 ==1 ]) #boolean indexing 
```

## slicing 

```python
arr = np.array([1,4,6,8,7,2])
print(arr[1:5])
print(arr[:6]) 
print(arr[3:]) 
print(arr[::2])
```
